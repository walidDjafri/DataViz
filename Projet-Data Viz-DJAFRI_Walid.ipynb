# Projet Data Viz : Analyse et visualisation des données d'un magasin le jour de  Black Friday

```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
import pandas
black_friday = pandas.read_csv("BlackFriday.csv")
```

```python
black_friday.head(8)
```

```python
black_friday.columns
```

```python
black_friday.info()
```

```python
Age_groupe = black_friday.groupby("Age")
```

```python
Age_groupe.count()
```

```python
achats_clients=dict()
for age,group in Age_groupe:
    achats_clients[age]=sum(group["Purchase"])
```

```python
achats_clients
```

```python
for age,group in Age_groupe:
    print(age)
    print(group)
```

```python
plt.plot(achats_clients.keys(), achats_clients.values(), marker='x')
plt.draw()
```

```python
plt.plot(achats_clients.keys(), achats_clients.values(), marker='o', linestyle="")
plt.xlabel("Catégorie d'âge")
plt.ylabel("Somme totale des ventes")
plt.title("Somme totale des ventes effectuées par catégorie d'âge", y=1.05)
plt.draw()
```

```python
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(achats_clients.keys(), achats_clients.values(), c=list(achats_clients.values()))
plt.xlabel("Catégorie d'âge")
plt.ylabel("Somme totale des ventes")
plt.title("Somme totale des ventes effectuées par catégorie d'âge", y=1.05)
plt.draw()
```

```python
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(achats_clients.keys(), achats_clients.values(), c=list(achats_clients.values()), cmap="seismic", marker="D", s=100)
plt.xlabel("Catégorie d'âge")
plt.ylabel("Somme totale des ventes")
plt.title("Somme totale des ventes effectuées par catégorie d'âge", y=1.05)
plt.colorbar()
plt.draw()
plt.savefig("mon_graphique.png", bbox_inches="tight")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# barplot avec la fonction sns.barplot représentant l'addition totale payée et en séparant ces données par sexe
sns.barplot(x="sex", y="total_bill", data=tips)
plt.draw()
```

```python
# la distribution de la variable total_bill grâce à la fonction distplot
sns.distplot(tips["total_bill"], kde=False)
plt.draw()
```

```python
# Utiliser la fonction jointplot de seaborn pour tracer la variable total_bill (X) en fonction de la variable tip (Y)
sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
# Utiliser la fonction jointplot de seaborn pour tracer la variable total_bill (X) en fonction de la variable tip (Y)
sns.jointplot(x="total_bill", y="tip", data=tips, kind='reg')
```

```python
# Utiliser la fonction pairplot de Seaborn pour étudier les relations entre nos variables
sns.pairplot(tips)
plt.draw()
```

```python
#  le boxplot de la variable total_bill en créant un boxplot par jour (day) 
# et en utilisant la variable sex dans l'option hue
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
plt.draw()
```
